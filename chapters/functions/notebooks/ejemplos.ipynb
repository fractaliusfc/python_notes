```python
def mrua5(x_i, v_i, a=-9.81):
    posiciones = []
    for i in range(0, 50, 1):
        t = i / 10
        x_f = x_i + v_i * t + 0.5 * a * t * t
        posiciones.append(x_f)
    return posiciones
```

```python
def mrua6(x_i, v_i, a=-9.81):
    posiciones = []
    tiempos = []
    for i in range(0, 50, 1):
        t = i / 10
        x_f = x_i + v_i * t + 0.5 * a * t * t
        posiciones.append(x_f)
        tiempos.append(t)
    return tiempos, posiciones
```

```python
resultado1, resultados2 = mrua6(30, 6)
```
